## 小範例

使用 [`matplotlib.pyplot.scatter( x, y, s, c, marker, linewidths, edgecolors='none' )`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) 函數可在直角坐標平面中繪製散佈圖

* 指令
  ```
  scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=<deprecated parameter>, edgecolors=None, *, plotnonfinite=False, data=None, kwargs)
  ```
* 參數說明  
  * 其他的樣式請參考 [matplotlib.markers](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers)

| 參數 | 意義 | 預設值 | 參數 |
|--------|:--------:|:--------:|:--------:|
| x, y | 各座標軸( x 軸與 y 軸 )維度的資料 |  |  |		
| s | 標記點的大小 |  |  |		
| c | 標記點的顏色 | blue(藍色) | r：紅色；b：藍色；g：綠色 |
| marker | 標記點的樣式 | 'o' | s：方塊；^：三角形；--：虛線；'*'：星號；'+' 加號 |
| cmap | 色彩 |  |  |		
| alpha | 標記點的透明度 |  | 參數範圍為 $[0,1]$，其中，0 表示完全透明，1 表示不透明 |	
| linewidths | 標記點邊緣線的寬度 | linewidths=1.5 |  |		
| edgecolor | 標記點邊緣線的顏色，用於空心符號 |  | 'face'：邊緣線與內部顏色相同；'none'(預設)：不繪製邊緣線 |	


````python
plt.style.use( 'default' )
# Data 
x = np.linspace( 0, 2*np.pi, num=80 )
y1 = np.cos( x )
y2 = np.sin( x )

# 散佈圖 Scatter Plot 
plt.scatter( x, y1, c='r', marker='*' )
plt.scatter( x, y2, c='b', marker='o' )
plt.show()
````

## 實作範例 - Iris Dataset

### 匯入資料

````python
from sklearn import datasets

# import some data to play with
iris = datasets.load_iris()    # 回傳格式為dictionary，將一個dict型別資料存入iris
````

````python
iris_feature = pd.DataFrame( iris['data'], columns=iris['feature_names'] )
iris_class = pd.DataFrame( iris['target'], columns=['target_names'] )

iris_data = pd.concat( [iris_feature, iris_class], axis=1)
iris_data.head()
````

### 基本繪圖方式

#### Matplot

````python
plt.style.use('default')

# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'])

# set a title and labels
#fig.suptitle('Iris Dataset')
ax.set_title('Scatter Plot for Iris Dataset')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
plt.show()
````

#### Seaborn

````python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=iris_data)
plt.title('Scatter Plot for Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
````

### 區分鳶尾花的種類

````python
fig, ax = plt.subplots()

scatter = ax.scatter( x='sepal length (cm)', y='sepal width (cm)', c='target_names', data=iris_data )

legend = ax.legend(*scatter.legend_elements(),
                    loc='upper right', title='Classes')
ax.add_artist(legend)

plt.show()
````

````python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target_names', data=iris_data)
plt.show()
````

### 調整顏色 & 型態

#### 顏色

````python
color = {
          0:'r',
          1:'g',
          2:'b'
}
iris_data['color'] = iris_data['target_names'].map(color)

iris_data.plot.scatter( x='sepal length (cm)', y='sepal width (cm)', c=iris_data['color'] )
plt.show()
````

#### 型態

````python
#hue表示按照Species對資料進行分類, 而style表示每個類別的標籤系列格式不一致.
sns.relplot(x='sepal length (cm)', y='sepal width (cm)', hue='target_names', style='target_names', data=iris_data )
plt.title('SepalLength and SepalWidth data by Target Names')
plt.show()
````

# Pairplot

* 可觀察每個情節的變化。繪圖採用矩陣格式，其中行名表示 x 軸，列名稱表示 y 軸
* 對角線圖是內核密度圖，其中其他圖是散點圖
* hue 在 sns.pairplot 函數調用中使用關鍵字
* 常用參數：
  * kind：用於控制非對角線 (左上至右下) 上的圖形，可選 "scatter" (散點圖) 與 "reg" (回歸直線)
  * diag_kind：用於控制對角線上的圖形，可選 "hist" (直方圖)與 "kde" (曲線圖)
