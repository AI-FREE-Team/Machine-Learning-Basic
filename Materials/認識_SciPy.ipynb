### 相對誤差指數函數

* `scipy.special.exprel(x)`：生成相對誤差指數，$\frac{exp(x) - 1}{x}$
* 當 x 接近 0 時，exp(x) 接近 1 ，因此exp(x)-1 的數值計算可能遭受災難性的精度損失。然後實踐 exprel(x)，以避免精度損失，這在 x 接近 0 時發生

````python
result = special.exprel([-0.25, -0.1, 0, 0.1, 0.25])
print(result) 
````

### 蘭伯特函數

* `scipy.special.lambertw` 也被稱為 Lambert W 函數。 
* Lambert W 函數 W(z) 被定義為 $w \times exp(w)$ 的反函數。換句話說，對於任何複數 z，W(z) 的值使得 $z=W(z) \times exp(W(z))$。
* Lambert W 函數是一個具有無限多個分支的多值函數。每個分支給出方程 $z=w \times exp(w)$ 單獨解。這裡，分支由整數 k 索引。

* 這裡的 Lambert W 函數是 $w \times exp(w)$ 的倒數。

````python
w = special.lambertw(1)
print( w )
print( w * np.exp(w) )
````

### 排列組合

* Combinations 組合函數的語法：`scipy.special.comb(N，k)`
* exact=False：大小寫接受數組參數。如果 N=0、k=0 則返回 0

````python
result = special.comb(10, 3, exact=False, repetition=True)
print( result )
````

* Permutations 組合函數的語法：`scipy.special.perm(N，k)`
* N 個事物排列一次取 k 個，即 N 的 k 排列，稱為「部分排列」

````python
result = special.perm(10, 3, exact = True)
print( result )
````

### 伽瑪函數

* 對於自然數 n，伽瑪函數通常被稱為廣義階乘，因為 $z \times gamma(z) = gamma(z + 1)$和 $gamma(n + 1)= n !$。
* 組合函數的語法：`scipy.special.gamma`。 N個事物排列取 k 個，即 N 的 k 排列，稱為「部分排列」。

````python
result = special.gamma([0, 0.5, 1, 5])
print( result )
````

### 計算方陣的行列式 `linalg.det`

### 取特徵值與特徵向量

* 矩陣(A)的特徵值(λ)、特徵向量(v)：$Av=λv$

````python
A = np.array([[3,4],[7,8]])
# 求解
l, v = linalg.eig(A)

print('特徵值；', l)
print('特征向量；\n', v)
````

### 計算方陣的逆矩陣 `linalg.inv`

````python
arr = np.array([[1, 2],
         [3, 4]])
iarr = linalg.inv(arr)
iarr
````

* 檢查

````python
np.allclose(np.dot(arr, iarr), np.eye(2))
````

* 奇異矩陣 (singular matrix) 不適用

````python
arr = np.array([[3, 2],
         [6, 4]])
linalg.inv(arr)
````

### 奇異值分解 (SVD) `linalg.svd`

SVD 常用於統計和信號處理，許多其他標準分解(QR、LU、Cholesky、Schur) 以及線性系統的求解器都在 [scipy.linalg](https://docs.scipy.org/doc/scipy/reference/linalg.html#module-scipy.linalg) 。

````python
arr = np.arange(9).reshape((3, 3)) + np.diag([1, 0, 1])
uarr, spec, vharr = linalg.svd(arr)
print('陣列頻譜: \n', spec)
````

* 檢查

````python
sarr = np.diag(spec)
svd_mat = uarr.dot(sarr).dot(vharr)
np.allclose(svd_mat, arr)
````

## 插值：[scipy.interpolate](https://docs.scipy.org/doc/scipy/reference/interpolate.html#module-scipy.interpolate)

[scipy.interpolate](https://docs.scipy.org/doc/scipy/reference/interpolate.html#module-scipy.interpolate) 可用於根據實驗數據擬合函數，進而評估不存在的點，該套件基於 [FITPACK Fortran subroutines](http://www.netlib.org/dierckx/index.html)。

### 建立線性的插值函數 `scipy.interpolate.interp1d`

````python
from scipy.interpolate import interp1d
````

````python
# 模擬正弦函數的實驗數據
measured_time = np.linspace(0, 1, 10)
noise = (np.random.random(10) * 2 - 1) * 1e-1
measures = np.sin(2 * np.pi * measured_time) + noise
````

````python
linear_interp = interp1d(measured_time, measures, kind='linear')
interpolation_time = np.linspace(0, 1, 50)
linear_results = linear_interp(interpolation_time)
````

* 通過提供 `kind` 可選的關鍵字參數，如：'Linear’、‘Nearest’、‘Zero’、‘Slinear’、‘Quadratic’、‘Cubic’等，下面案例選擇三次插值：

````python
cubic_interp = interp1d(measured_time, measures, kind='cubic')
cubic_results = cubic_interp(interpolation_time)
````

* 方法一

````python
plt.scatter(measured_time, measures, label='measures')
plt.plot(interpolation_time, linear_results, color='red', label='linear interp')
plt.plot(interpolation_time, cubic_results, color='green', label='cubic interp')
plt.legend()
plt.show()
````

* 方法二

````python
plt.plot(measured_time, measures, 'o',
     interpolation_time, linear_results, '-',
     interpolation_time, cubic_results, '--')
plt.legend(['measures', 'linear interp', 'cubic interp'], loc='best')
plt.show()
````

`scipy.interpolate.interp2d` 類似於 `scipy.interpolate.interp1d`，適用於二維數組，要留意：對於 interp 系列，插值點必須保持在給定數據點的範圍內。有關深入的插值範例，請參閱 [Maximum wind speed prediction at the Sprogø station](https://scipy-lectures.org/intro/summary-exercises/stats-interpolate.html#summary-exercise-stat-interp)。

### 噪聲數據插值

* 透過 interpolate 套件中 UnivariateSpline 對含有噪音的數據進行插值運算。
* 使用 UnivariateSpline ，輸入一組數據點，通過繪製一條平滑曲線來去除噪聲。繪製曲線十，可設置平滑參數 s，如果參數 s=0，將對所有點(含噪聲)進行插值運算，也就是說 s=0 時不去除噪聲。

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
````

````python
x = np.linspace(-3, 3, 50)
y = np.exp(-x**2) + 0.1 * np.random.randn(50) # 通过random方法添加噪声数据

plt.plot(x, y, 'ro', ms=5)

# 平滑参数使用默认值
spl = UnivariateSpline(x, y)
xs = np.linspace(-3, 3, 1000)
plt.plot(xs, spl(xs), 'b', lw=3) # 蓝色曲线

# 设置平滑参数
spl.set_smoothing_factor(0.5)
plt.plot(xs, spl(xs), 'g', lw=3) # 绿色曲线

# 设置平滑参数为0
spl.set_smoothing_factor(0)
plt.plot(xs, spl(xs), 'yellow', lw=3) # 黄色曲线

plt.show() 
````

## 優化與適配：[scipy.optimize](https://docs.scipy.org/doc/scipy/reference/optimize.html#module-scipy.optimize)

### 曲線擬合 `optimize.curve_fit`

#### 單變數

* 關於正弦波的數據，其中包含一些噪聲：

````python
# 產生含噪聲的樣本數據
x_data = np.linspace(-5, 5, num=50)                   # 採樣點
y_data = 2.9 * np.sin(1.5 * x_data) + np.random.normal(size=50)   # 加入隨機樹作為噪聲
````

如果知道數據位於正弦波上，而不是振幅或週期，可以通過最小二次曲線擬合找到這些數據。首先，我們必須定義要擬合的測試函數，這裡是幅度和周期未知的正弦：

````python
def test_func(x, a, b):
    return a * np.sin(b * x)
    
params, params_covariance = optimize.curve_fit(test_func, x_data, y_data, p0=[2, 2])
# 輸出結果
print('求出的係數 a, b: ')  # 给定模型的最優參數
print(params)

print('\nvariables_covariance: ')  # 可用於檢查擬合情況，其對角線元素值表示每個參數的方差
print(params_covariance)
````

````python
y = test_func(x_data, params[0], params[1])

plt.plot(x_data, y_data, 'o', color='green', label='Samples')
plt.plot(x_data, y, color='red', label='Fit')
plt.legend(loc='best')

plt.show()
````

#### 多變數

* 最小二乘法
  * optimize 套件提供實現最小二乘擬合算法的函數`leastsq()`，`leastsq` 是 least square 的簡寫，即最小二乘法。
  * 語法
  ```
  optimize.leastsq(func, x0, args=())
  ```
  * 參數
    * `func`：計算誤差的函數
    * `x0`：計算的初始參數值
    * `args`：指定 func 的其他參數

````python
# 樣本數據
X = np.array([160,165,158,172,159,176,160,162,171])
Y = np.array([58,63,57,65,62,66,58,59,62])

# 偏差函數，計算以 p 為參數的直線和原始數據之間的誤差
def residuals(p):
        k, b = p
        return Y-(k*X+b)

# leastsq() 使得 residuals() 的輸出數組的平方法最小，參數的最初始值為 [1, 0]
result = optimize.leastsq(residuals, [1, 10])
k, b = result[0]
print('k = ', k, 'b = ', b)
````

````python
# 畫樣本點
plt.figure(figsize=(8, 6))     ## 指定圖像比例： 8：6
plt.scatter(X, Y, color='green', label='Samples', linewidth=2)

# 畫擬合直線
x = np.linspace(150, 190, 100) ##在150-190直接劃100個連續點
y = k*x + b ##函數式
plt.plot(x, y, color='red', label='Fit', linewidth=2)
plt.legend()   #繪圖
plt.show()
````

````python
#Rosenbrock Function
def fun_rosenbrock(x):
   return np.array([10 * (x[1] - x[0]**2), (1 - x[0])])
from scipy.optimize import least_squares
input = np.array([2, 2])
result = optimize.least_squares(fun_rosenbrock, input)
print('確切的最小值：', result.x)
````

### 尋找函數的最小值 `optimize.minimize`

````python
from scipy.optimize import minimize
````

#### 單變數

````python
# 目標函數
def f(x):
    return x**2 + 10*np.sin(x)
````

````python
x = np.arange(-10, 10, 0.1)
plt.plot(x, f(x)) 
plt.axhline(y=0, color='red')
plt.show() 
````

該函數的全局最小值約為 -1.3，局部最小值約為 3.8，可使用 尋找最小值 `scipy.optimize.minimize()`，給定起點 x0，它返回它找到的最小值的位置，`scipy.optimize.minimize()` 是一個複合對象，包含關於收斂的所有信息：

````python
# 設定參數、求解與結果
result = minimize(f, x0=0)
print(result)
print('='*20) 
print(result.x)
````

#### 單變數帶限制式

* 方法一：使用 `bounds` 參數將變量約束到區間：

````python
result = minimize(f, x0=1, bounds=((-5,10), ))
result.x
````

* 方法二：`optimize.fminbound`

* 要求取一定範圍之内的函数最小值，可使用 `fminbound` 方法，可得：函數最小值是-7.945823，對應的 x 值為-1.306440

````python
# 求取-5到10之间的函数最小值。full_output=True 表示返回详细信息。
result = optimize.fminbound(f, -5, 10, full_output=True)
print(result)
````

#### 多變數

* 解 $f(x,y)=x^3-4xy+2y^2$ 相對極小值

````python
# 目標函數
def f(x, sign=1):
        x1 = x[0]
        x2 = x[1]
        return sign*(x1**3-4*x1*x2 +2*x2**2)
````

````python
# 設定參數
x0 = [1,1]    # 起始值(initial values)

# 求解與結果
Result4 = minimize(f, x0)
print( '解出的x值：', Result4.x )
print( '解出的極小值:', Result4.fun ) 
print( 'f(Result1.x):', f(Result4.x) )
````

另一解 (0, 0) 是鞍點，如果設定 x0=[0,0]，則會得到極值：0。因為鞍點判斷的程式作法需要賦予更多的條件。如果想挑戰 Python 程式，把目前所學過的方法串起來，可以循以下步驟：
* 步驟 1：使用 diff() 函數解一階偏微分，求取可能的臨界值
* 步驟 2：求二階偏微分
* 步驟 3：參考主題一的二元一次方程式求解判斷的準則 (Delta)，定義 Hessian 來判斷誰是鞍點、誰無解
* 最後：把可能的臨界值設為起始值、求解

#### 多變數帶限制式

* 解 $Min. x+2y (x>0, y>0)$ s.t. $xy=5000$

````python
# 目標函數
def f(x, sign=1):
        x1 = x[0]
        x2 = x[1]
        return sign*(x1 + 2*x2)
````

````python
# 限制條件
def constraint1(x, sign=1):
        return sign*(x[0]*x[1] - 5000)
````

````python
# 設定參數
x0 = [10,10]     # 起始值，
b1 = (0, np.inf)   # 參數條件x>0，上界給予正無限大，np.inf就是¥
b2 = (0, np.inf)   # 參數條件y>0，上界給予正無限大，np.inf就是¥
bnds= (b1, b2)     # 邊界條件向量
con1 = {'type':'ineq', 'fun':constraint1} # 把限制集定義成字典
cons = [con1]     # 把con1做成串列 (萬一有多個條件時，可以包在一起)
````

````python
# 求解與結果
Result5 = minimize(f, x0,bounds=bnds,constraints=cons)
print( '解出的x值：', Result5.x )
print( '解出的極小值:', Result5.fun ) 
print( 'f(Result1.x):', f(Result5.x) )
````

## 快速傅立葉變換：[scipy.fftpack](https://docs.scipy.org/doc/scipy/reference/fftpack.html#module-scipy.fftpack)

[scipy.fftpack](https://docs.scipy.org/doc/scipy/reference/fftpack.html#module-scipy.fftpack) 套件計算快速傅立葉變換 (FFT) 並提供實用程序來處理它們，主要功能是：
* `scipy.fftpack.fft()` 計算 FFT
* `scipy.fftpack.fftfreq()` 生成採樣頻率
* `scipy.fftpack.ifft()` 計算逆 FFT，從頻率空間到信號空間

````python
from scipy import fftpack
````

### 產生信號

````python
# Seed the random number generator
np.random.seed(1234)

time_step = 0.02
period = 5.

time_vec = np.arange(0, 20, time_step)
sig = (np.sin(2 * np.pi / period * time_vec)
       + 0.5 * np.random.randn(time_vec.size))

plt.figure(figsize=(6, 5))
plt.plot(time_vec, sig, label='Original signal')
plt.show()
````

### 計算和繪製功率
  * 由於信號來自實函數，因此傅立葉變換是對稱的
  * 峰值信號頻率可以用 `freqs[power.argmax()]`
  * 將高於此頻率的傅立葉分量設置為零並使用 對 FFT 進行反相`scipy.fftpack.ifft()`，得到一個濾波信號

````python
# The FFT of the signal
sig_fft = fftpack.fft(sig)

# And the power (sig_fft is of complex dtype)
power = np.abs(sig_fft)**2

# The corresponding frequencies
sample_freq = fftpack.fftfreq(sig.size, d=time_step)

# Plot the FFT power
plt.figure(figsize=(6, 5))
plt.plot(sample_freq, power)
plt.xlabel('Frequency [Hz]')
plt.ylabel('plower')

# Find the peak frequency: we can focus on only the positive frequencies
pos_mask = np.where(sample_freq > 0)
freqs = sample_freq[pos_mask]
peak_freq = freqs[power[pos_mask].argmax()]

# Check that it does indeed correspond to the frequency that we generate
# the signal with
np.allclose(peak_freq, 1./period)

# An inner plot to show the peak frequency
axes = plt.axes([0.55, 0.3, 0.3, 0.5])
plt.title('Peak frequency')
plt.plot(freqs[:8], power[:8])
plt.setp(axes, yticks=[])
plt.show()
# scipy.signal.find_peaks_cwt can also be used for more advanced
# peak detection
````

### 去除所有高頻

````python
high_freq_fft = sig_fft.copy()
high_freq_fft[np.abs(sample_freq) > peak_freq] = 0
filtered_sig = fftpack.ifft(high_freq_fft)

plt.figure(figsize=(6, 5))
plt.plot(time_vec, sig, label='Original signal')
plt.plot(time_vec, filtered_sig, linewidth=3, label='Filtered signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.legend(loc='best')
plt.show()
````

### 邊緣檢測

* 是一種尋找圖像中物體邊界的圖像處理技術，原理是通過檢測圖像中的亮度突變，來辨識物體邊緣。
* 用在圖像處理、計算機視覺、機器視覺等領域中廣泛應用。
* 常用邊緣檢測算法：
  * Sobel
  * Canny
  * Prewitt
  * Roberts
  * Fuzzy Logic methods

````python
im = np.zeros((256, 256))
im[64:-64, 64:-64] = 1
im[90:-90, 90:-90] = 2
im = ndimage.gaussian_filter(im, 8)

import matplotlib.pyplot as plt
plt.imshow(im)
plt.show()
````

圖像看起來像一個正方形的色塊，將檢測這些彩色塊的邊緣。這裡使用 ndimage 的 Sobel 函數來檢測圖像邊緣，該函數會對圖像數組的每個軸分開操作，產生兩個矩陣，然後使用 NumPy 中的 Hypot 函數將這兩個矩陣合併為一個矩陣，得到最後結果。

````python
im = np.zeros((256, 256))
im[64:-64, 64:-64] = 1
im[90:-90, 90:-90] = 2
im = ndimage.gaussian_filter(im, 8)

sx = ndimage.sobel(im, axis=0, mode='constant')
sy = ndimage.sobel(im, axis=1, mode='constant')
sob = np.hypot(sx, sy)

plt.imshow(sob)
plt.show()
````
