本篇文章將說明機器學習中基本的專業用語，藉此理解機器學習模型的建立與評量過程。讀者們在閱讀本篇文章後，將具備下列基礎知識：

# 機器學習的種類  

根據用來訓練模型的資料，我們可將機器學習分成 3 大類 : 
- <font color="#00dd00">**監督式學習( Supervised Learning )**</font> :   
> 資料中會提供問題的答案( 即標籤 )，讓模型來學習資料中的特徵( Features )與標籤( Label )的關聯性。  
- <font color="#00dd00">**非監督式學習( Unsupervised Learning )**</font> :  
> 資料中沒有提供任何問題的標籤，讓模型自己探索出資料中的關聯性。  
- <font color="#00dd00">**強化學習( Reinforcement Learning )**</font> :  
> 環境中代理人( Agent )學習獲得最大獎勵的過程，如同一位電玩玩家，在遊戲中不斷觀察與嘗試，以獲取高分獲得勝利。

# <font color="#00dd00">**迴歸( Regression )**</font>與<font color="#00dd00">**分類( Classification )**</font>  

在機器學習的監督式學習( Supervised Learning )演算法中，主要在解決兩類問題 : <font color="#00dd00">迴歸( Regression )</font>及<font color="#00dd00">分類( Classification )</font>。<font color="#dddd00">**迴歸**問題係透過演算法來預測一個實數值</font>，例如 : 預測明年的銷售量、預測新冠肺炎確診人數等；<font color="#dddd00">**分類**問題則係透過演算法來預測屬於哪一種類別</font>，當預測的類別只有兩種結果時，我們稱此為一個二元分類( Two-class Classification )問題，例如 : 明年顧客會不會續約、明天股市漲或跌、照片中有沒有狗等，而當類別有三種以上的結果時，我們稱其為一個多元分類( Multi-class Classification )問題，例如 : 辨識手寫數字 0~9、歸納新聞類別、判斷照片中的動物等。實務上，當一個專案在設定假說時，我們需先釐清解決問題的類型，再選用適當的演算法來解決問題。

 

# <font color="#00dd00">**損失函數( Loss Function )**</font>  
 
在監督式學習的任務中，<font color="##00dd00">**損失函數( Loss Function )**</font><font color="#dddd00">是用來度量模型的 $\hat{y_i}$ 預測值( Prediction )與 $y_i$ 實際值( Ground Truth )之間的差異程度</font>，並透過<font color="#00dd00">**最佳化問題**</font>尋找損失函數**最小值**的過程，來獲得模型最終的參數。  
有關損失函數的定義，在迴歸問題中，我們希望預測值與實際值的「誤差( Error )」或稱「殘差( Residual )」越小越好，故損失函數定義成預測值與實際值的誤差總和，另一方面，在分類問題中，我們則希望預測類別的錯誤率越小越好，換句話說，希望模型預測正確類別的機率越大越好，因此會藉由模型對各個類別的預測機率來定義損失函數。



在迴歸問題中，基本且常見的損失函數有 :  
  1. [**平均絕對誤差( Mean Absolute Error, MAE )**](https://en.wikipedia.org/wiki/Mean_absolute_error)  
$\displaystyle \text{MAE} = \frac{\displaystyle \sum_{i=1}^N \lvert \hat{y_i}-y_i \rvert}{N} \text{ ，又稱為 } L_1 \text{ 損失；}$  
</br>
  2. [**均方誤差( Mean Sequare Error, MSE )**](https://en.wikipedia.org/wiki/Mean_squared_error)  
$\displaystyle \text{MSE} = \frac{\displaystyle \sum_{i=1}^N ( \hat{y_i}-y_i )^2}{N} \text{ ，又稱為 } L_2 \text{ 損失；}$  
</br>   
  3. [**均方根誤差( Root Mean Sequare Error, RMSE )**](https://en.wikipedia.org/wiki/Root-mean-square_deviation)  
$\displaystyle \text{RMSE} = \sqrt{\frac{\displaystyle \sum_{i=1}^N ( \hat{y_i}-y_i )^2}{N}} \text{ ； }$  
</br>   
  4. [**平均絕對百分比誤差( Mean Absolute Percentage Error, MAPE )**](https://en.wikipedia.org/wiki/Mean_squared_error)  
$\displaystyle \text{MAPE} = \frac{1}{N} \sum_{i=1}^N \bigg| \frac{y_i-\hat{y_i}}{y_i} \bigg| \text{ ； }$  
</br>   

在分類問題中，常用的損失函數為 :    
  1. [**交叉熵誤差( Cross Entropy Error )**](https://en.wikipedia.org/wiki/Cross_entropy)  
$\displaystyle \text{E} = -\displaystyle \sum_{c=1}^C \sum_{i=1}^N y_{c,i} \ln(\hat{y}_{c,i}) , \\ \text{其中， } C \text{ 為總類別數，} N \text{ 為資料總筆數，}y_{c,i} \text{ 以 One-Hot 形式表示每筆資料所屬實際的類別標籤( Label )，} \hat{y}_{c,i} \text{ 表示每筆資料預測所屬類別的機率。}$  
</br>

# 評估方法   

在監督式學習的任務中，須根據問題的類型，選擇適當的評估指標，常見的評估指標如下 :  
<br/>


|迴歸問題|分類問題|  
|:---:|:---:|
|平均絕對誤差( Mean Absolute Error )|混淆矩陣( Confusion Matrix )|
|均方誤差( Mean Square Error )| 正確率( Accuracy )|
|決定係數( Coefficient of Determination )|精確率( Precision )|
||召回率( Recall )|
||F1-Score|
||Area Under the Curve( AUC )|  

## 迴歸問題  

關於迴歸問題的評估指標，我們除可透過損失函數，例如 : 平均絕對誤差( MAE )、均方誤差( MSE )等，來檢視模型的預測表現，還有另一個常見的評估指標 - $R^2$ 決定係數( Coefficient of Determination )。

### [決定係數( Coefficient of Determination )](https://en.wikipedia.org/wiki/Coefficient_of_determination)   

在統計學中，決定係數( Coefficient of Determination )標記為 $R^2$ ，常用來判別 $X$ 自變數與 $Y$ 應變數是否存有線性的關聯性。其數學定義如下 :  
</br>   
\begin{equation}
\begin{split}
\displaystyle R^2 &= 1-\frac{\displaystyle \sum_{i=1}^N ( \hat{y_i}-y_i )^2 }{\displaystyle \sum_{i=1}^N ( \hat{y_i}-\overline{y} )^2 } \text{ ，其中 } \overline{y} \text{ 為真實資料的平均值；} \\
& = 1 - \frac{MSE}{Var(Y)} \text{ 。 }
\end{split}
\end{equation}
</br>   

一般而言，$R^2$ 決定係數是一個介於 0 到 1 的實數，<font color="#dddd00">當 $R^2$ 越接近 1 時，表示迴歸模型對資料的解釋能力越好。</font> 此外，從上列 $R^2$ 的數學定義中，我們可將 $R^2$ 視為利用均方誤差( MSE )來評估模型的表現，然而，因均方誤差容易受資料的離散程度所影響，因此，當目標變數的分佈區間範圍落差較大時，通常我們會選用 $R^2$ 作為評估指標。  
<br/>
註 : 更多關於 $R^2$ 與 MSE 的介紹與說明，有興趣的讀者們可參考 [Mean Squared Error or R-Squared – Which one to use ?](https://vitalflux.com/mean-square-error-r-squared-which-one-to-use/)。


## 分類問題

### [混淆矩陣( Confusion Matrix )](https://en.wikipedia.org/wiki/Confusion_matrix)  

在分類問題中，混淆矩陣係藉由表格來呈現模型的分類結果，以預測結果與實際標籤構成二維資料表，明確掌握分類錯誤的情況。下圖為二元分類的混淆矩陣( 圖片來源 : [Confusion Matrix for Your Multi-Class Machine Learning Model](https://towardsdatascience.com/confusion-matrix-for-your-multi-class-machine-learning-model-ff9aa3bf7826) ) : 

<p align="center">
<img width="550" src="https://raw.githubusercontent.com/AI-FREE-Team/Machine-Learning-Basic/main/Content/Confusion_Matrix.png">
</p>





二元分類的混淆矩陣，將由預測結果與實際標籤的對照，組成下列四種情況，分別為 :  
* <font color="#00dd00">True Positive( TP )</font> : 實際為正，預測為正；   
* <font color="#00dd00">True Negative( TN )</font> : 實際為負，預測為負；  
* <font color="#00dd00">False Positive( FP )</font> : 實際為負，預測為正；    
* <font color="#00dd00">False Negative( FN )</font> : 實際為正，預測為負。  
</br>  

進一步，我們利用上述混淆矩陣中的四種情況，計算出下列數值，作為評估模型的指標 :    
  - <font color="#00dd00">**正確率( Accuracy )**</font>：模型正確預測的比率   
>  $\displaystyle Accuracy = \frac{ TP + TN }{ TP + FP + TN + FN } $   
    
  - <font color="#00dd00">**精確率( Precision )**</font>：在所有模型預測為正類的資料中，實際也為正類的資料比率
> $\displaystyle Precision = \frac{ TP }{ TP + FP } $   
    
  - <font color="#00dd00">**召回率( Recall )**</font>：在所有實際為正類的資料中，模型能正確預測為正類的資料比例      
> $\displaystyle Recall = \frac{ TP }{ TP + FN } $    

  註 : <font color="#dddd00">精確率與召回率呈**反向**關係</font>，也就是說，一方提升時，另一方就會下降。 
  
  - <font color="#00dd00">**特異度( Specificity )**</font>：在所有實際為負類的資料中，模型能正確預測為負類的資料比例   
> $\displaystyle Specificity = \frac{ TN }{ TN + FP } $    

  - <font color="#00dd00">**F1 分數( F1 Score )**</font>：Precision 與 Recall 的調和平均  
> $\displaystyle \frac{ 1 }{ F1 } = \frac{ 1 }{ Precision } + \frac{ 1 }{ Recall } $   

上列指標的選用，應依據問題的目的而有所不同，舉例來說，若我們想要透過模型來檢測病患狀況是否為惡性，召回率將會是一個重要的評量指標，表示實際為惡性的病患，能透過模型準確地研判出來，另外同樣的邏輯概念，在商品瑕疵檢測的過程中，我們則希望能透過模型偵測出任何的瑕疵商品。

### [接收者操作特徵( Receiver Operating Characteristic, ROC )](https://en.wikipedia.org/wiki/Receiver_operating_characteristic)  

在分類問題中，另一個重要的評量指標 - <font color="#00dd00">**接收者操作特徵( ROC )**</font>。在機器學習的專案中，當資料有嚴重**不平衡( imbalanced )**的狀況時，上述混淆矩陣中所介紹的各種評估指標，將無法有效評估模型的表現。因此，我們將再進一步藉由 **ROC 曲線** 與 **AUC( Area Under the Curve )**的概念，來評估模型在不平衡資料集中的表現。





ROC 曲線主要由下列兩個指標所構成 :  
- <font color="#00dd00">False Positive Rate( FPR )</font> : 模型預測為正類但實際為負類的資料，佔所有負類資料的比例   
> $\displaystyle FPR = \frac{FP}{FP+TN} $
- <font color="#00dd00">True Positive Rate( TPR )</font> : 模型預測為正類且實際也為正類的資料，其佔所有正類資料的比例   
> $\displaystyle TPR = \frac{TP}{TP+FN} $  

在 ROC 曲線的圖中，橫軸( x 座標軸 )表示 FPR ， FPR 越大代表模型預測正類的資料中，實際為負類的資料越多，另外，縱軸( y 座標軸 )為 TPR， TPR 越大代表模型預測正類的資料中，實際也為正類的資料越多。因此，在最理想的狀態下，我們希望 FPR 越小越靠近 0 越好，且 TPR 越大越靠近 1 越好，故<font color="#dddd00"> ROC 曲線越靠近 $(0,1)$ 點時，表示模型分類的表現越佳</font>，如下圖所示( 圖片來源 : [Wikipedia](https://en.wikipedia.org/wiki/Receiver_operating_characteristic) &nbsp;) :

# <font color="#00dd00">**過擬合( Overfitting )**</font>與<font color="#00dd00">**欠擬合( Underfitting )**</font>    

在機器學習中，我們的終極目的是訓練出一個具泛化能力( Generalization )的模型，換句話說，模型在前所未見的資料中也能有良好的表現。因此，在監督式學習的任務中，為了希望模型具有普遍的適用性，當我們在訓練模型時，會將資料集分割成下列三個子集 : 
- 訓練集( Training Set ) : 用來訓練模型；
- 驗證集( Validation Set ) : 用來評量模型的訓練成效，我們可透過模型在驗證集中的表現，回過頭來調整模型的超參數( Hyperparamters )，使模型能有更好的表現；
- 測試集( Testing Set ) : 用來檢視模型在訓練與驗證完成後的表現，透過測試集的評估，我們將能得到前述介紹的評估指標，例如：決定係數( Coefficient of Determination )、正確率( Accuracy )等。 

一般而言，我們會依據資料量的大小，來分配每個子集的比例，在過去傳統機器學習的年代，若資料量達到萬筆等級，通常以訓練集 60%、驗證集 20%、測試集 20% 為比例原則，而在現今大數據的時代，通常我們則以 98%/ 1%/ 1% 的比例做分割。

<br/>  
  
|資料量(N) &emsp;&emsp;&emsp;&emsp;&nbsp;&emsp;&emsp; |訓練集|驗證集|測試集|
|---|---:|---:|---:|
|$N \geq 100,000$|60%|20%|20%|
|$N \geq 1,000,000$|98% <br/> 99.5%|1% <br/> 0.5%|1% <br/> 0.1%|

<br/>
   
註 : 有關資料集分配比例的原則，請參閱 [Train / Dev / Test sets](https://www.coursera.org/lecture/deep-neural-network/train-dev-test-sets-cxG1s)、[Splitting a Dataset into Train and Test Sets](https://www.baeldung.com/cs/train-test-datasets-ratio)。   
  
<br/>
   
當模型在驗證集中的表現遠遠不及在訓練集中的表現時，這種狀況我們稱 - <font color="#00dd00">**過擬合( Overfitting )**</font>，代表模型學習能力過強，能捕捉到訓練集中所有樣本的特性，但也可能因此學習到訓練集中的噪音，且將這些噪音認為是一般的通則，而導致在實際應用時預測的失準。另一方面，<font color="#00dd00">**欠擬合( Underfitting )**</font>則代表模型的學習能力過於薄弱，無法充份捕捉到資料的特性，而導致泛化能力的不足。

關於如何防止過擬合、欠擬合的方法，下面將列舉一些常見方法的關鍵字，供讀者們延伸學習時參考：
- 過擬合
  - 增加訓練集的資料量  
  - 交叉驗證( Cross Validation )  
  - 特徵選取( Feature Selection )   
  - 正則化( Regularization )
  - 集成學習( Ensemble Learning )  
- 欠擬合  
  - 增加新特徵  
  - 降低正則化的參數
  - 使用更複雜的演算法或模型
